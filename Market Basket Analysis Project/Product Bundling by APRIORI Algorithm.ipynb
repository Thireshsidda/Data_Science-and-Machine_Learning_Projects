### Step 1: Importing the required libraries

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

### Step 2: Loading and exploring the data

```python
#Exploring the Different sub Classifications of transactions
df['Sub Classification'].unique()
```

## Step 3: Cleaning the Data

```python
# Stripping extra charecters in the Item Description
df['Item Description']=df['Item Description'].str.strip('')
```

```python
df['Item Description']
```

```python
#Dropping the rows without any Project Code
df.dropna(axis=0 , subset=['Project Code'], inplace= True)
df['Project Code']=df['Project Code'].astype('str')
```

## Step 4: Splitting the data according to the Sub classification of transaction

```python
df['Sub Classification'].value_counts()
```

```python
#Transaction done on 'Adult' Sub Classifcation
basket_Adult=(df[df['Sub Classification']=='Adult']
              .groupby(['Project Code','Item Description'])['Unit Price']
              .sum().unstack().reset_index().fillna(0)
              .set_index('Project Code'))
basket_Adult
```

## Step 5:Hot encoding the Data

```python
#Defining the hot encoding function to make the data suitable for the concerned libraries.
def hot_encode(x):
    if(x<=0):
        return False
    if(x>=0):
        return True
```

```python
basket_encoded=basket_Adult.applymap(hot_encode)
basket_encoded
```

## Step 6: Building the models and analyzing the results

```python
#Building the model
frq_items = apriori(basket_encoded, min_support = 0.05, use_colnames= True)

#Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric='lift', min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
rules.head()
```
